# Imported Libraries

```python
import cv2
import numpy as np
import os
import shutil
from tensorflow.keras import callbacks
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import Dense, Flatten, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import IPython.display as display
from PIL import Image
import time
from sklearn.metrics import classification_report, confusion_matrix
```

# Setup for Face Detection

Needed Parameters

```python
CONFIDENCE_THRESHOLD = 0.5
input_shape = (224, 224, 3)
num_classes = 105  # Assuming there are 105 unique celebrities in the dataset
```

UINT8 Model for Transfer Learning

```python
model_file = 'D:/noody/Deep learning/Lab 4&5/opencv_face_detector_uint8.pb'
config_file = 'D:/noody/Deep learning/Lab 4&5/opencv_face_detector.pbtxt'
net = cv2.dnn.readNetFromTensorflow(model_file, config_file)
```

# Load Dataset and Apply Face Detection

Detect and Crop Faces

```python
def detect_and_crop_face(image):
    h, w = image.shape[:2]
    blob = cv2.dnn.blobFromImage(image, 1.0, (300, 300), (104.0, 177.0, 123.0))
    net.setInput(blob)
    detections = net.forward()

    cropped_faces = []
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        if confidence > CONFIDENCE_THRESHOLD:
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (startX, startY, endX, endY) = box.astype("int")
            face = image[startY:endY, startX:endX]
            cropped_faces.append(face)
    return cropped_faces
```

Process Dataset and Save Cropped Faces

```python
def process_dataset(dataset_path, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for celebrity_folder in os.listdir(dataset_path):
        folder_path = os.path.join(dataset_path, celebrity_folder)
        if os.path.isdir(folder_path):
            celebrity_output_folder = os.path.join(output_folder, celebrity_folder)
            if not os.path.exists(celebrity_output_folder):
                os.makedirs(celebrity_output_folder)
            for img_name in os.listdir(folder_path):
                img_path = os.path.join(folder_path, img_name)
                image = cv2.imread(img_path)
                if image is None:
                    print(f"Warning: Unable to load image {img_path}. Skipping this file.")
                    continue
                cropped_faces = detect_and_crop_face(image)
                for idx, face in enumerate(cropped_faces):
                    face_resized = cv2.resize(face, (224, 224))  # Resize face to input shape
                    face_name = f"{img_name.split('.')[0]}_face{idx}.jpg"
                    cv2.imwrite(os.path.join(celebrity_output_folder, face_name), face_resized)
```

Dataset paths

```python
dataset_path = 'D:/noody/Deep learning/Lab 4&5/105_classes_pins_dataset'
output_folder = 'D:/noody/Deep learning/Lab 4&5/105_classes_pins_dataset_cropped'
process_dataset(dataset_path, output_folder)
```

Load and Split Data

```python
def load_dataset(dataset_folder):
    X, y = [], []
    label_map = {}  # Map to store string labels to integer labels
    current_label_id = 0
    
    for celebrity_folder in os.listdir(dataset_folder):
        folder_path = os.path.join(dataset_folder, celebrity_folder)
        if celebrity_folder not in label_map:
            label_map[celebrity_folder] = current_label_id
            current_label_id += 1
        label = label_map[celebrity_folder]
        
        for img_name in os.listdir(folder_path):
            img_path = os.path.join(folder_path, img_name)
            image = cv2.imread(img_path)
            if image is not None:
                X.append(image)
                y.append(label)
    
    return np.array(X), np.array(y), label_map

# Load dataset and apply label encoding
X, y, label_map = load_dataset(output_folder)
```

Train, Validate, and Test Split for Images

```python
def split_and_save_images(dataset_path, output_base_dir, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15):
    # Create directories for train, val, and test
    for split in ['train', 'validation', 'test']:
        split_dir = os.path.join(output_base_dir, split)
        if not os.path.exists(split_dir):
            os.makedirs(split_dir)

    for celebrity_folder in os.listdir(dataset_path):
        folder_path = os.path.join(dataset_path, celebrity_folder)
        if os.path.isdir(folder_path):
            # Get all image files
            images = [img for img in os.listdir(folder_path) if img.endswith(('jpg', 'jpeg', 'png'))]
            
            # If fewer than 3 images, place all images in the train set
            if len(images) < 3:
                print(f"Folder {celebrity_folder} has fewer than 3 images. Adding all to training set.")
                train_imgs = images
                val_imgs, test_imgs = [], []  # No validation/test images
            else:
                # Split the images into train, val, and test
                train_imgs, temp_imgs = train_test_split(images, test_size=(val_ratio + test_ratio), random_state=42)
                val_imgs, test_imgs = train_test_split(temp_imgs, test_size=test_ratio / (val_ratio + test_ratio), random_state=42)

            # Define destination directories
            split_dirs = {
                'train': os.path.join(output_base_dir, 'train', celebrity_folder),
                'validation': os.path.join(output_base_dir, 'validation', celebrity_folder),
                'test': os.path.join(output_base_dir, 'test', celebrity_folder)
            }
            
            for split, imgs in zip(['train', 'validation', 'test'], [train_imgs, val_imgs, test_imgs]):
                # Create class subdirectory if it doesn't exist
                if not os.path.exists(split_dirs[split]):
                    os.makedirs(split_dirs[split])
                
                # Copy each image to the corresponding directory
                for img in imgs:
                    src_path = os.path.join(folder_path, img)
                    dst_path = os.path.join(split_dirs[split], img)
                    shutil.copyfile(src_path, dst_path)
    
    print("Dataset split into train, validation, and test directories.")

# Specify the input and output directories
output_base_dir = 'D:/noody/Deep learning/Lab 4&5/105_classes_pins_dataset_split'
split_and_save_images(output_folder, output_base_dir)
```

# Pre-trained Model for Face Recognition

InceptionV3 Model

```python
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=input_shape)

# Add classification layers on top of it
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(2048, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)

predictions = Dense(num_classes, activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=predictions)
```

Freeze Base Model Layers

```python
for layer in base_model.layers:
    layer.trainable = False
```

Compile Model

```python
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

Data Augmentation

```python
# Set up paths to the train, validation, and test directories
train_dir = os.path.join(output_base_dir, 'train')
validation_dir = os.path.join(output_base_dir, 'validation')
test_dir = os.path.join(output_base_dir, 'test')

# Create the ImageDataGenerator objects
train_datagen = ImageDataGenerator(rescale=1.0/255.0, rotation_range=20, width_shift_range=0.2,
                                   height_shift_range=0.2, horizontal_flip=True)
val_datagen = ImageDataGenerator(rescale=1.0/255.0)
test_datagen = ImageDataGenerator(rescale=1.0/255.0)

# Create the data generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    validation_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```

Train Model

```python
# Define early stopping callback
early_stopping = callbacks.EarlyStopping(
    monitor='val_loss',  # Monitor validation loss
    patience=3,           # Number of epochs with no improvement after which training will be stopped
    restore_best_weights=True  # Restore model weights from the epoch with the best value of the monitored quantity
)

history = model.fit(train_generator, validation_data=val_generator, epochs=25, callbacks=[early_stopping])
```

Evaluate on Test Data

```python
test_loss, test_accuracy = model.evaluate(test_generator)

print(f"Test Accuracy: {test_accuracy * 100:.2f}%")  
```

Confusion Matrix

```python
# Obtain predictions on the test data
y_pred = model.predict(test_generator)
y_pred_labels = np.argmax(y_pred, axis=1)  # Convert probabilities to predicted class labels

# True labels from the test generator
y_true = test_generator.classes

# Print confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred_labels)
print("Confusion Matrix:")
print(conf_matrix)
```

Evaluation Metrics

```python
# Define target names for the classes that are actually in y_true
unique_labels = np.unique(y_true)  # Get unique classes in y_true
target_names = [name for i, name in enumerate(test_generator.class_indices.keys()) if i in unique_labels]

# Generate a classification report with precision, recall, and F1-score
class_report = classification_report(y_true, y_pred_labels, labels=unique_labels, target_names=target_names)
print("Classification Report:")
print(class_report)
```

Visualize Validation and Test Accuracy and Loss

```python
# Store test accuracy and loss in the history object
history.history['test_loss'] = test_loss
history.history['test_accuracy'] = test_accuracy

# Plot validation & testing accuracy and loss curves
def plot_validation_and_testing(history):
    # Retrieve validation and test metrics
    val_accuracy = history.history['val_accuracy']
    test_accuracy = [history.history['test_accuracy']] * len(val_accuracy)
    val_loss = history.history['val_loss']
    test_loss = [history.history['test_loss']] * len(val_loss)
    epochs = range(1, len(val_accuracy) + 1)

    # Plot accuracy
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, val_accuracy, 'bo-', label='Validation Accuracy')
    plt.plot(epochs, test_accuracy, 'go-', label='Test Accuracy')
    plt.title('Validation and Test Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs, val_loss, 'bo-', label='Validation Loss')
    plt.plot(epochs, test_loss, 'go-', label='Test Loss')
    plt.title('Validation and Test Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Call the function to plot the validation and testing history
plot_validation_and_testing(history)
```

Feature Extractor Model

```python
feature_extractor = Model(inputs=model.input, outputs=model.layers[-2].output)  # Remove the final classification layer
```

Feature Extraction from Image

```python
def get_embedding(image):
    # Preprocess and resize image
    image = cv2.resize(image, (224, 224))
    image = image.astype("float32") / 255.0
    image = np.expand_dims(image, axis=0)
    return feature_extractor.predict(image).flatten()  # Flatten the output
```

Calculate and Store Class Embeddings

```python
# Build class embeddings with flattened embeddings
class_embeddings = {}
for class_label in os.listdir(train_dir):
    class_folder = os.path.join(train_dir, class_label)
    embeddings = []
    
    for img_name in os.listdir(class_folder)[:10]:  # Use first 10 images as samples
        img_path = os.path.join(class_folder, img_name)
        image = cv2.imread(img_path)
        embedding = get_embedding(image)
        embeddings.append(embedding)
    
    # Calculate the average embedding for each class and flatten it
    class_embeddings[class_label] = np.mean(embeddings, axis=0)
```

Identification Based on Cosine Similarity

```python
def is_identified(embedding, threshold=0.5):
    similarities = [
        cosine_similarity([embedding], [known_emb])[0][0]  # Pass as 2D array
        for known_emb in class_embeddings.values()
    ]
    max_similarity = max(similarities)
    return max_similarity > threshold, max_similarity
```

Classify Image Using Known Embeddings

```python
def classify_image(image):
    embedding = get_embedding(image)
    identified, confidence = is_identified(embedding)
    if not identified:
        return "not identified"
    
    # Compute similarities again to find the most similar class
    similarities = [
        cosine_similarity([embedding], [known_emb])[0][0]
        for known_emb in class_embeddings.values()
    ]
    most_similar_class = list(class_embeddings.keys())[np.argmax(similarities)]
    return most_similar_class, confidence
```

Classify Faces in Live Camera Feed

```python
def classify_live_feed_jupyter():
    cap = cv2.VideoCapture(0)
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # Detect faces and classify each face here
        faces = detect_and_crop_face(frame)
        for face in faces:
            face_resized = cv2.resize(face, (224, 224))
            label, confidence = classify_image(face_resized)
            label_text = f"{label} ({confidence:.2f})" if label != "not identified" else label
            cv2.putText(frame, label_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # Display the frame in Jupyter Notebook
        _, img = cv2.imencode('.png', frame)
        display.clear_output(wait=True)
        display.display(Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)))
        time.sleep(0.1)  # Control the update rate for smoother output

    cap.release()

classify_live_feed_jupyter()
```
